```python
#fuction to count number of black pieces and kings
def count_black(grid):
    piece=0;
    king =0;
    for x in grid:
        for y in x:
            if(y=="b"):
                piece+=1
            if(y=="bk"):
                king+=1
    return piece,king;
#function to count number of red pieces and kings
def count_red(grid):
    piece=0;
    king=0;
    for x in grid:
        for y in x:
            if(y=="r"):
                piece+=1
            if(y=="rk"):
                king+=1
    return piece,king
```

```python
#convert the black pieces to kings if in last row of red and same for red pieces
def convert_king(grid):
    
    for j in range(0,8):
        if(grid[7][j]=="b"):
            grid[7][j]="bk"
        if(grid[0][j]=="r"):
            grid[0][j]="rk"
    return grid
```

```python

def print_grid(grid):
    for x in grid:
        print(x);
    print("\n")
```

```python


def model(w0,w1,w2,w3,w4,w5,w6):
    
    e=10
    a=0.1
    grid=[["g","b","g","b","g","b","g","b"],
      ["b","g","b","g","b","g","b","g"],
      ["g","b","g","b","g","b","g","b"],
      ["g","g","g","g","g","g","g","g"],
      ["g","g","g","g","g","g","g","g"],
      ["r","g","r","g","r","g","r","g"],
      ["g","r","g","r","g","r","g","r"],
      ["r","g","r","g","r","g","r","g"],
     ]
    possible_states=[]
    x1,x3=count_black(grid)
    x2,x4=count_red(grid)
    x5=threaten_black(grid)
    x6=threaten_red(grid)
    prevy=(w0+w1*x1+w2*x2+w3*x3+w4*x4+w5*x5+w6*x6)%100;
    nxtst=grid
    while(e):
        convert_king(grid)
        err=0
        yopt=-1;
        for state in possible_states:
            x1,x3=count_black(state)
            x2,x4=count_red(state)
            x5=threaten_black(state)
            x6=threaten_red(state)
            score=(w0+w1*x1+w2*x2+w3*x3+w4*x4+w5*x5+w6*x6)%100
            if(yopt<score):
                yopt=score
                nxtst=state
        
        if(yopt!=-1):
            dlta=(prevy-yopt);
            w1=w1+a*dlta
            w2=w2+a*dlta
            w3=w3+a*dlta
            w4=w4+a*dlta
            w5=w5+a*dlta
            w6=w6+a*dlta
        grid=nxtst
        possible_stb=possible_black(grid)
        #print(possible_stb[0])
        grid=possible_stb[0]
        possible_states=possible_red(grid)
        convert_king(grid)
        e-=1
    print("Optmal grid state after 100 epochs:-\n");
    print_grid(grid);
    print("Optimal Weights after the 100 epochs:-\n");
    print(f"w0={w0}\n")
    print(f"w1={w1}\n")
    print(f"w2={w2}\n")
    print(f"w3={w3}\n")
    print(f"w4={w4}\n")
    print(f"w5={w5}\n")
    print(f"w6={w6}\n")
```
