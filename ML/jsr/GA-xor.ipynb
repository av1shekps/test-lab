**Q) Implement a genetic algorithm program to successfully classify examples in the 2 input EXOR problem using Gabilâ€™s method.**

```python
import numpy as np
import pandas as pd
import random
```

```python
n = 100
population = []
for i in range(n):
  s = ''
  a = np.random.rand(12)*10
  for x in a:
    if (int)(x) % 2 == 0:
      s += '0'
    else:
      s += '1'
    
  population.append(s)

print(population)
```

```python
def _XOR(inp1, inp2, out):
  if (inp1 == inp2 and out == '0') or (inp1 != inp2 and out == '1'):
    return 1
  else:
    return 0
```

```python
def fitness(s):
  o=0;
  if s[0]=='0' and s[1]=='0' and s[2]=='0':
    o+=0.25
  if s[3]=='0' and s[4]=='1' and s[5]=='1':
    o+=0.25
  if s[6]=='1' and s[7]=='0' and s[8]=='1':
    o+=0.25
  if s[9]=='1' and s[10]=='1' and s[11]=='0':
    o+=0.25
  return o*o
```

```python
curr_fitness = 0
epoch = 0

while curr_fitness < 1:
  ps = []
  fit_list = []

  for x in population:
    fitness_score = fitness(x)
    fit_list.append([fitness_score,x])

  fit_list.sort()
  for i in range(60, 100):
      ps.append(fit_list[i][1])
  
  for i in range(30):  
    l1=list(fit_list[i][1])
    l2=list(fit_list[59-i][1])
    cp1 = random.randrange(12) % 12
    cp2 = random.randrange(12) % 12
    nl1=l1[:cp1]+l2[cp1:cp2]+l1[cp2:12]
    nl2=l2[:cp1]+l1[cp1:cp2]+l2[cp2:12]
    ps.append(''.join(nl1))
    ps.append(''.join(nl2))

  mutation = ps[60]
  mut_ind = random.randrange(12) % 12
  if mutation[mut_ind] == '0':
    c = '1'
  else:
    c = '0'

  mutated = mutation[:mut_ind] + c + mutation[mut_ind + 1:]
  
  ps[60] = mutated

  population = ps
  curr_fitness=fit_list[-1][0]
  epoch+=1
  print("epoch : " + str(epoch) + " fitness : " + str(curr_fitness))
```
