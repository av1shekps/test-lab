**Q) Implement a Genetic algorithm to generate solutions for 8-Queens problem.**

```python
import numpy as np
import pandas as pd
import random
```

```python
def checkRowClashes(chromosome):
  cnt = 0
  for i in range(len(chromosome)):
    for j in range(i+1, len(chromosome)):
      if chromosome[i] == chromosome[j]:
        cnt = cnt + 1
  
  return cnt
```

```python
def checkDiagonalClashes(chromosome):
  cnt = 0
  for i in range(len(chromosome)):
    for j in range(i+1, len(chromosome)):
      if abs(chromosome[j] - chromosome[i]) == abs(j - i):
        cnt += 1
  
  return cnt
```

```python
def calculateFitness(chromosome):
  fitness = 0
  fitness += checkRowClashes(chromosome)
  fitness += checkDiagonalClashes(chromosome)
  # print("Fitness: ", fitness)
  return fitness
```

```python
p = []
n = 8
for i in range(n):
  chromosome = []
  for j in range(8):
    gene = random.randrange(8) % 8
    chromosome.append(gene)
  p.append(chromosome)
```

```python
curr_fitness = 100
fitness_threshold = 0
epoch = 0
best_fitness = 100
while curr_fitness > fitness_threshold and epoch < 5000:
  fit_list = []
  for chromosome in p:
    fit_score = calculateFitness(chromosome)
    fit_list.append([fit_score, chromosome])
  
  fit_list.sort()
  # print(fit_list)
  curr_fitness = fit_list[0][0]
  if curr_fitness < best_fitness:
    best_fitness = curr_fitness
    ans = fit_list[0][1]
  print("After epoch " + str(epoch) + " , fitness = " + str(best_fitness))
  crossover_element1 = fit_list[0][1]
  crossover_element2 = fit_list[1][1]

  crossover_point = random.randrange(8) % 8
  child1 = crossover_element1[:crossover_point] + crossover_element2[crossover_point:] 
  child2 = crossover_element2[:crossover_point] + crossover_element1[crossover_point:] 
  
  child1[random.randrange(8) % 8] = random.randrange(8) % 8
  child2[random.randrange(8) % 8] = random.randrange(8) % 8

  ps = []
  ps.append(child1)
  ps.append(child2)
  for i in range(2, len(fit_list)):
    ps.append(fit_list[i][1])

  p = ps
  epoch += 1
```

```python
def print_board(ans):
  board = []
  for i in range(8):
    row = []
    for j in range(8):
      row.append('*')
    board.append(row)

  for i in range(8):
    board[ans[i]][i] = 'Q'

  for i in range(8):
    print(board[i])
```

```python
print("After 5000 epochs, best possible solution: ")
print(ans)
print_board(ans)
```
