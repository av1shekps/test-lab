**Q) Design and implement machine learning algorithm using least means square learning rule to play checkers game.**

```python
import numpy as np
import copy
import random
```

```python
#Size of the board
size = 8 
```

```python
def createBoard():
    board = [[' ', 'B', ' ', 'B', ' ', 'B', ' ', 'B'],
         ['B', ' ', 'B', ' ', 'B', ' ', 'B', ' '],
         [' ', 'B', ' ', 'B', ' ', 'B', ' ', 'B'],
         [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
         [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
         ['R', ' ', 'R', ' ', 'R', ' ', 'R', ' '],
         [' ', 'R', ' ', 'R', ' ', 'R', ' ', 'R'],
         ['R', ' ', 'R', ' ', 'R', ' ', 'R', ' ']]
    return board
```

```python
def possibleJump(board,symbol,x1,y1,x2,y2,x3,y3):
    if(x3 < 0 or x3 >= size or y3 < 0 or y3 >= size):
        return False
    if(board[x3][y3]!=' '):
        return False
    if (symbol == 'R' or symbol == 'KR'):
        if (board[x1][y1] == 'R' and x3 > x1):
            return False
        if (board[x2][y2] != 'B' and board[x2][y2] != 'KB'):
            return False
        return True;
    elif(symbol == 'B' or symbol == 'KB'):
        if (board[x1][y1] == 'B' and x3 < x1):
            return False
        if (board[x2][y2] != 'R' and board[x2][y2] != 'KR'):
            return False
        return True
```

```python
def possibleMove(board,symbol,x1,y1,x2,y2):
    if (x2 < 0 or x2 >= size or y2 < 0 or y2 >= size):
        return False
    if (board[x2][y2] != ' '):
        return False
    if (symbol == 'R' or symbol == 'KR'):
        if (board[x1][y1] == 'R' and x2 > x1):
            return False
        return True
    elif(symbol == 'B' or symbol == 'KB'):
        if (board[x1][y1] == 'B' and x2 < x1):
            return False
        return True
```

```python
def endGame(board, symbol):
    for i in range(size):
        for j in range(size):
            if(board[i][j]!=symbol):
                return 0
    return 1
```

```python
def initKings(board):
  for j in range(size):
    if(board[0][j]=='R'):
      board[0][j]='KR'
    if(board[size-1][j]=='B'):
      board[size-1][j]='KB'
```

```python
def getCloneForJump(boardState, r1, c1, r2, c2, r3, c3):
  clone = copy.deepcopy(boardState)
  clone[r3][c3]=clone[r1][c1]
  clone[r1][c1]=' '
  clone[r2][c2]=' '
  initKings(boardState)
  return clone
```

```python
def getCloneForMove(boardState, r1, c1, r2, c2):
  clone = copy.deepcopy(boardState)
  clone[r2][c2]=clone[r1][c1]
  clone[r1][c1]=' '
  initKings(boardState)
  return clone
```

```python
def getCorrectMoves(boardState, coin):
    
    CorrectMoves = []
    
    if(coin=='B'):
        kp='KB'
    else:
        kp='KR'
    for i in range(len(boardState[0])):
        for j in range(len(boardState[0])):
            if(boardState[i][j]==coin or boardState[i][j]==kp):
                if(possibleJump(boardState,coin,i,j,i+1,j-1,i+2,j-2)):
                    CorrectMoves.append(getCloneForJump(boardState, i, j, i+1, j-1, i+2, j-2))
                if(possibleJump(boardState,coin,i,j,i-1,j-1,i-2,j-2)):
                    CorrectMoves.append(getCloneForJump(boardState, i, j, i-1, j-1, i-2, j-2))
                if(possibleJump(boardState,coin,i,j,i+1,j+1,i+2,j+2)):
                    CorrectMoves.append(getCloneForJump(boardState, i, j, i+1, j+1, i+2, j+2))
                if(possibleJump(boardState,coin,i,j,i-1,j+1,i-2,j+2)):
                    CorrectMoves.append(getCloneForJump(boardState, i, j, i-1, j+1, i-2, j+2))
    if(len(CorrectMoves)==0):
      for i in range(len(boardState[0])):
            for j in range(len(boardState[0])):
                if(boardState[i][j]==coin or boardState[i][j]==kp):
                    if(possibleMove(boardState,coin,i,j,i+1,j-1)):
                        CorrectMoves.append(getCloneForMove(boardState, i, j, i+1, j-1))
                    if(possibleMove(boardState,coin,i,j,i-1,j-1)):
                        CorrectMoves.append(getCloneForMove(boardState, i, j, i-1, j-1))
                    if(possibleMove(boardState,coin,i,j,i+1,j+1)):
                        CorrectMoves.append(getCloneForMove(boardState, i, j, i+1, j+1))
                    if(possibleMove(boardState,coin,i,j,i-1,j+1)):
                        CorrectMoves.append(getCloneForMove(boardState, i, j, i-1, j+1))
    
    return CorrectMoves
```

```python
def getFeaturesX(board, s1, s2):
    
    feature_x = np.zeros(7)
    feature_x[0]=1
    for i in range(size):
        for j in range(size):
            if(board[i][j]==s1):
                feature_x[1]+=1
            elif(board[i][j]==s2):
                feature_x[2]+=1
            elif(board[i][j]==('K'+s1)):
                feature_x[3]+=1
            elif(board[i][j]==('K'+s2)):
                feature_x[4]+=1
            if(board[i][j]==s1 or board[i][j]==('K'+s1)):
                if(possibleJump(board,s1,i,j,i+1,j+1,i+2,j+2)):
                    feature_x[5]+=1
                if(possibleJump(board,s1,i,j,i-1,j+1,i-2,j+2)):
                    feature_x[5]+=1
                if(possibleJump(board,s1,i,j,i+1,j-1,i+2,j-2)):
                    feature_x[5]+=1
                if(possibleJump(board,s1,i,j,i-1,j-1,i-2,j-2)):
                    feature_x[5]+=1
            elif(board[i][j]==s2 or board[i][j]==('K'+s2)):
                if(possibleJump(board,s2,i,j,i+1,j+1,i+2,j+2)):
                    feature_x[6]+=1
                if(possibleJump(board,s2,i,j,i-1,j+1,i-2,j+2)):
                    feature_x[6]+=1
                if(possibleJump(board,s2,i,j,i+1,j-1,i+2,j-2)):
                    feature_x[6]+=1
                if(possibleJump(board,s2,i,j,i-1,j-1,i-2,j-2)):
                    feature_x[6]+=1
    return feature_x
```

```python
def getNonFinalBoardScore(weights,features):                                    # Computes weighted sum of weight vector and feature vector to get the Non-Final Board state
    weights = np.array(weights).reshape((len(weights),1))
    features = np.array(features).reshape((len(features),1))
    boardScore = np.dot(weights.T,features)
    return(boardScore[0][0])
```

```python
def chooseMoveWithMaxBoardScore(board,s1,s2,weights):                           # Chooses the next move based on the board state having maximum board score

    rightMoves = getCorrectMoves(board,s1)
    if(len(rightMoves)==0):
      return -1
    rightMovescores = [getNonFinalBoardScore(weights,
                                             getFeaturesX(i,s1,s2)) for i in rightMoves]
    return rightMoves[np.argmax(rightMovescores)]
```

```python
def chooseRandomMove(board,s):                                                  # For computer to learn by doing random moves
    legalMoves = getCorrectMoves(board,s)
    if(len(legalMoves)==0):
      return -1
    return random.choice(legalMoves)
```

Opponent chooses random moves, with which game results are stored and will be used for training



```python
def trainModel(coins,w1,w2,board):
    gameResults = []
    gameStatusFlag = True
    
    clone = copy.deepcopy(board)
    while(gameStatusFlag):
        clone = chooseMoveWithMaxBoardScore(clone,coins[0],coins[1],w1)
        print("Player chooses move: ", clone)
        if(clone==-1):
          break
        gameResults.append(clone)
        gameStatusFlag = not endGame(clone,coins[0])
        if(gameStatusFlag == False):
            break
        clone = chooseRandomMove(clone,coins[1])
        if(clone==-1):
          break;
        gameResults.append(clone)
        gameStatusFlag =  not endGame(clone,coins[1])
    return gameResults
```

```python
def getFinalBoardScore(board,s1,s2):

    # If draw
    score = 0
    # If computer wins
    if(endGame(board,s1)):
        score = 100
    # If opponent wins
    elif(endGame(board,s2)):
        score = -100
    
    return score
```

```python
def getTrainingExamples(weights,s1,s2,trainData):

    trainingData=[]
    for i in range(len(trainData)-1):
        features = getFeaturesX(trainData[i+1],s1,s2)
        trainingData.append([features,getNonFinalBoardScore(weights,features)])
    trainingData.append([getFeaturesX(trainData[-1],s1,s2),
        getFinalBoardScore(trainData[-1],s1,s2)])
    return trainingData
```

```python
def getgameResult(s1,s2,gameResult,trainData):

    finalScore = getFinalBoardScore(trainData[-1],s1,s2)
    if(finalScore == 100):
        gameResult[0] += 1
    elif(finalScore == -100):
        gameResult[1] += 1
    else:
        gameResult[2] += 1
    return gameResult
```

```python
def LMS(weights,trainData,alpha=0.01):
    for train in trainData:
        boardStateTrain = train[1]
        boardStateScore = getNonFinalBoardScore(weights,train[0])
        weights = weights + (alpha * (boardStateTrain - boardStateScore) * np.array(train[0]))
    return weights
```

```python
coins = ('R','B')
numTrainingExamples = 15
```

Training the Model \\

```python
weights = [np.random.rand(7),np.random.rand(7)]
gameResults = [0,0,0]
i = 0
for i in range(numTrainingExamples):
    startingBoardState = createBoard()
    trainingData = trainModel(coins,weights[0],weights[1],startingBoardState)
    PlayerOne = getTrainingExamples(weights[0],coins[0],coins[1],trainingData)
    PlayerTwo = getTrainingExamples(weights[1],coins[1],coins[0],trainingData)
    gameResults = getgameResult(coins[0],coins[1],gameResults,trainingData)
    weights = [LMS(weights[0],PlayerOne),LMS(weights[1],PlayerOne)]
print("After Training: (" + "Player 1 won : " + str(gameResults[0]) +
    ", Player 2 won : " + str(gameResults[1]) + ", Draws : " + str(gameResults[2]) +
    ")\n")
learntWeight =  list(np.mean(np.array([weights[0],weights[1]]),axis = 0))
print("Weight Vector obtained: \n"+ str(learntWeight))
```
